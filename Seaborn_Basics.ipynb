```python
# Load the necessary libraries
import pandas as pd
import seaborn as sns
sns.set(style="darkgrid") # This isn't something you have to do, this is like a style template
```

```python
# Read the CSV file with customer churn data.
url = 'http://bit.ly/comm_churn'
df_churn = pd.read_csv(url)
```

```python
# Use Pandas to get an idea of what the data looks like
df_churn.head(5)
```

```python
# DISTRIBUTION PLOTS
sns.displot(df_churn, x="monthly_charges"); # So in this example, how many customers (y axis) are at what amount of bill (x axis)
```

```python
# Here's another way to show this data using kde or kernel density estimation
sns.displot(df_churn, x="monthly_charges", kind="kde"); #Looks similar to the last plot, but has the density on the y axis
```

```python
sns.displot(df_churn, x="monthly_charges", kind="kde", fill=True); # Purely aesthetic
```

```python
sns.displot(df_churn, x="monthly_charges", hue="internet_service");  # hue is for color, but in this case it's going to fill in color based on kind of internet service
```

```python
sns.displot(df_churn, x="monthly_charges", hue="internet_service", kind="kde", fill=True); # Combine it all
```

```python
# COUNT PLOTS
sns.countplot(x="contract_type", data=df_churn); # This will count categories, this will show the amount of customers with each contract type, in this example
```

```python
sns.countplot(x="contract_type", hue="churn", data=df_churn);  # So we're going to color based on whether or not they churn
# In this example, we see that customers that only have one option a year, or every two years, to leave typically don't. vs that of month to month customers, who have the option to leave more often
```

```python
sns.countplot(y="contract_type", hue="churn", data=df_churn); # Simply turn the graph sideways, note that it's now y= instead of x=
```

```python
sns.countplot(y="contract_type", hue="churn",palette="Set2", data=df_churn); # Literally just changing to a preset color scheme, located in Seaborn documentation 
```

```python
# BOX PLOTS
sns.boxplot(x="payment_method", y="monthly_charges", data=df_churn);
# Line in the middle of the box is the median
# Size of the box is the variation/range of the data
```

```python
sns.boxplot(x="phone_service", y="monthly_charges", data=df_churn);
# So customers that have phone service have a higher monthly median than not
```

```python
sns.boxplot(x="phone_service", y="monthly_charges", hue="senior_citizen", data=df_churn);
# For example here, you can see senior citizens with phon service have less variation in their monthly bills
```

```python
# Read the CSV file with ad spend and revenue.
url = 'http://bit.ly/ad_spend'
df_ads = pd.read_csv(url)
```

```python
df_ads.head()
```

```python
# SCATTER PLOTS
sns.scatterplot(x="TV", y="sales", data=df_ads);
# Scatter plots show us the relationship between two variables
# It seems that as spending on TV ads goes up, so does sales.  REMEMBER CORRELATION ISN'T CAUSATION
```

```python
sns.scatterplot(x="TV", y="sales", size="radio", data=df_ads);
# Same data, but also with the amount of radio ad spend happening during a given campaign
```

```python
sns.scatterplot(x="TV", y="sales", size="radio", hue="radio", data=df_ads);
# Same story, but with the colors, it's easier to visualize
```

```python
# PAIR PLOT
sns.pairplot(df_ads) ;
# Gives you a smattering of charts, perhaps giving you an idea on things to further investigate
```

```python
sns.pairplot(df_ads.iloc[:, [1,2]]) ;
# We're pulling a specific part of the dataset to generate that slice of the data.
# Rememeber we can only feed in numerical data into these, so it'll throw an error if there's nonnumerical data involved (like the yes/no for phone service in the previous)
```
