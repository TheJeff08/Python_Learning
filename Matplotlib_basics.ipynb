#### Basic Graph

```python
# Great place to start is always the Matplotlib documentation https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.html
# You can ctrl f to look for things that you'd want to use
plt.plot([1,2,3], [2,4,6])  # Here you'll want to remember to add the 'plt.' to the plot function becasue we imported matplotlib as plt
```

```python
# To make things easier to manage it's best to move our arrays into variables:
x = [1,2,3]
y = [2,4,6]
plt.plot(x, y)
```

```python
# As you can see Jupyter Notebook is showing the lines thing above the graph, you can get rid of that by doing as follows:
plt.plot(x, y)
plt.show()
```

```python
# From here we can go back to the documentation to see what else we can do to our basic graph.
# For example if we wanted to set a title, we can find these things.
# The below is expanding from the video lesson

x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x, y)

plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis!')
plt.ylabel('Y Axis!')

plt.xticks([0,1,2,3,4]) # These set the numbers exactly of what the ticks on the axes are.  You can do even integers, you can do decimals, and it will scale and place each as is necessary
plt.yticks([0,2,4,6,8,10])

plt.legend()  # This breaks it because there's no label included in the original plot.  We'll need to pass through a label along with the plot, see below

plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x, y, label='2x', color='green', linewidth='3', linestyle='--', marker='.', markersize='10', markeredgecolor='orange' ) # Passing in colors, you can use a hex value for example, too

# There's a shorthand notation that will let you do style things quickly
# fmt = '[color][marker][line]' SEE NEXT GRAPH FOR EXAMPLE ON THIS LINE

plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis!')
plt.ylabel('Y Axis!')

plt.xticks([0,1,2,3,4]) 
plt.yticks([0,2,4,6,8,10])

plt.legend()  

plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# plt.plot(x, y, label='2x', color='green', linewidth='3', linestyle='--', marker='.', markersize='10', markeredgecolor='orange' ) 

# There's a shorthand notation that will let you do style things quickly
# fmt = '[color][marker][line]' 

plt.plot(x, y, 'g.-', label='2x')  # In the documentation there's a whole list of markers you can use in shorthand

plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis!')
plt.ylabel('Y Axis!')

plt.xticks([0,1,2,3,4]) 
plt.yticks([0,2,4,6,8,10])

plt.legend()  

plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]


plt.plot(x, y, 'g.-', label='2x') 

## Line number Two
x2 = np.arange(0,4.5,0.5)
print(x2)   # Note that this gives you a NumPy array on top of the graph
plt.plot(x2[:5], x2[:5]**2, 'r', label='x^2')
plt.plot(x2[4:], x2[4:]**2, 'r--')   # showing projections past a point, having the line graphed to a certain point, 4 in this case, and then dashed past that point in the second line of code.

plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis!')
plt.ylabel('Y Axis!')

plt.xticks([0,1,2,3,4]) 
#plt.yticks([0,2,4,6,8,10])  # Since x2 was higher than the original plot, we commented out the yticks here to see what would default, and it looks fine.

plt.legend()  

plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

## Resize your graph
plt.figure(figsize=(5,3), dpi= 300) # This will make an arbitrary 5x3 x to y, you can also pass in dpi after the figsize to specific ppi, he recommends around 300

plt.plot(x, y, 'g.-', label='2x') 

## Line number Two
x2 = np.arange(0,4.5,0.5)
print(x2)
plt.plot(x2[:5], x2[:5]**2, 'r', label='x^2')
plt.plot(x2[4:], x2[4:]**2, 'r--') 

plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis!')
plt.ylabel('Y Axis!')

plt.xticks([0,1,2,3,4]) 


plt.legend()  

plt.savefig('mygraph.png', dpi=300) # For example, you can set the dpi when you save the graph, so it'll pass to the saved image

plt.show()
```

#### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

bars[0].set_hatch('/') # Check out A bar
bars[1].set_hatch('O') 
bars[2].set_hatch('*')

plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

# Let's say we didn't want to write out each bar ---
#bars[0].set_hatch('/') # Check out A bar
#bars[1].set_hatch('O') 
#bars[2].set_hatch('*')

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.show()
```
