#### Filtering Rows
##### df.loc[] and df.iloc[]

```python
# df.iloc[] will use the index location, and df.loc[] will use the names
```

```python
df.head()
```

```python
# Say we wanted the age of the person in row 2, Allison, Miss. Helen Loraine
df.iloc[2, 4] # Keeping in mind counting starts at 0, and in this example the header columns and the 0,1,2,3,4 on the side don't count
```

```python
# You can also provide a single value, which would then just filter to the row
df.iloc[5]
```

```python
# If we put it into a list, we'll filter to row 5 as a DataFrame
df.iloc[[5]]
```

```python
# Similarly, if we wanted to filter just a column, we can do df.iloc[:, x] where x is the column we wanted, and then putting it in a list returns it as a DataFrame
```

```python
# If we wanted to do this with loc, we'd need to put in the names, for example
df.loc[:, ['pclass', 'survived']]
```

#### Filtering Rows

```python
# Can use boolean expressions to filter
df['age'] > 5
```

```python
# Returns whether or not the person that died was over 5 years old.  You can then pass this into df.loc and it'll filter it down to whenever the person was over 5
df.loc[df['age'] > 5]
```

```python
# A useful thing to note is that you can do the inverse of any boolean expression by adding a ~ to the front of it, so let's return everyone except those older than 5
df.loc[
    ~(df['age'] > 5)    # Note that the expression was grouped in () for this to work
]
```

```python
# Use query to do this similarly
df.query('pclass == 3 and age > 5')  # So note the use of () here and also that I'm using boolean to return two conditions
```
