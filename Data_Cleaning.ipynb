```python
import pandas as pd
```

```python
df = pd.read_excel(r"E:\Learning_files\titanic3.xls")
df
```

```python
# Good first thing to do is remove your duplicates, this will quickly clean up any duplicates, then update the value of df
df = df.drop_duplicates()
df
```

```python
# From here it can be useful to clean up the data by eliminating any columns that aren't necessary for your use case

df = df.drop(columns = ["body", "sibsp", "parch"])
df
```

```python
# Now examine the data to see how we can make the data better/remove nulls/make more better
df["name"].str.rstrip(")")     # This will only be useful if the value is before/after.  For example if the / in home.dest was at the front or back, here we might need to use replace
                                # For here what we'll do is show an example of just pulling the ) out of the row 4 name using rstrip.  lstrip is a thing as well
```

```python
# As you can see this applied it only to the singular column, and if we were to refresh df to this value, df will become only that column of info.
# What we want to do is reset the name column to this new value without that ) in there

df["name"] = df["name"].str.rstrip(")")  # Again, we're just replacing that given column with the function that removed that unwanted character
df
```

```python
# The next example deals with splitting the address, some of which have just a street, others with other info like ZIP or whatever.  We'll do this live with the titanic set and names

df["name"].str.rsplit(',',n=1, expand=True) # We're telling it to split on the comma, and then how many values to the right to split on, in this case 1, and we'll see how that does
```

```python
# Because there was only one comma, that was enough to split where we wanted it to, so now we can rename the columns
df[["Last_Name", "First_Name"]] = df["name"].str.rsplit(',',n=1, expand=True)
df
```

```python
# We can further break this down into a suffix column
df["First_Name"].str.split('.',n=1, expand=True)
```

```python
# Likewise, re add them into the rull DataFrame
df[["Suffix", "First_Name"]] = df["First_Name"].str.split('.',n=1, expand=True)
df
```

```python

# To the end of this (because we can use where I stopped for tomorrow's project)

# We've been looking column by column for things to clean or manipulate.  We can now look at the whole sheet and, for example, replace null values entirely (the NaN)

df = df.fillna('')
df

# Essentially this is going to look over the entire DataFrame, and replace all null values with nothing
# From here, we can likewise look for any other common n/a type entries that we'd want to remove and do the same as the above
# Thinking back to the Phone_Number cleaning, there were a bunch of folks on the Do not call list, and there were some that didn't have a phone number listed
# We'll use the drop function, but first we have to define the elements that determine what is dropped

for x in df.index:
	if df.loc[x, "Do_Not_Contact"] == 'Y':
	df.drop(x, inplace=True)
	
df
# So, looking at the above, we're using loc which we learned about in the Pandas basic notebook, we're eliminating the index anywhere where the customer is marked 'Y' in the Do_Not_Contact column

for x in df.index:
	if df.loc[x, "Phone_Number"] == '':
	df.drop(x, inplace=True)
	
df
# So, likewise here, we looked into the Phone_Number column and anything that was blank, we eliminated, because obviously we can't call them.
#Another way to tackle the blank phone number problem would be to drop null values, that would be done as the following:
df = df.dropna((subset="Phone_Number"), inplace=True)
# In the video, the data is nice and clean and almost ready for the client, but the index is all messed up, with the numbers jumping around because of what we've eliminated

df.reset_index(drop=True)  # This will reset the index values, and the drop=True inside the () makes it so it doesn't generate a copy of the previous index and list it there.
# Finally, we can set the above = to df, and we've got a nice clean DataFrame for the client

Merging DataFrames

Merge_DataFrame.ipynb![image.png](attachment:b80b5517-2d92-4eb4-bb6d-a6407243638a.png)![image.png](attachment:e60aa152-6b65-4e95-845f-d2aa25b8d668.png)![image.png](attachment:35e9c4b4-ef85-4808-8bd6-a87f0686b1b4.png)
```
